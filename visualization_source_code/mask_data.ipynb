```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_USA = ['April', 'May', 'June']

Y_USA = [69.6, 81.8, 86.1]

X_USA2 = ['April', 'May', 'June']

Y_USA2 = [74.7, 83.1, 86.4]

X_USA3 = ['April', 'May', 'June']

Y_USA3 = [79.8, 80.7, 88.3]

X_USA4 = ['April', 'May', 'June']

Y_USA4 = [83.7, 84.7, 92.4]


colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361']

plt.figure(figsize = (10.5, 5))
plt.plot(X_USA, Y_USA, color = '#003f5c')
plt.plot(X_USA2, Y_USA2, color = '#58508d')
plt.plot(X_USA3, Y_USA3, color = '#bc5090')
plt.plot(X_USA4, Y_USA4, color = '#ff6361')
plt.legend(['18-29', '30-44', '45-59', '60+'], loc=4)
plt.ylabel('% reported wearing face mask')
plt.xlabel('Month in 2020')
plt.title('Percent of People in Different Age Brackets who Reported Face Mask Use from April to June 2020 (USA)')
plt.show()
```

```python

plt.figure(figsize = (10.5, 7))
height_always = [98.9, 95.7 ]
bars = ['Women', 'Men']
fig, ax = plt.subplots()
ax.bar(bars, height = height_always, label = 'always', color = '#003f5c' )
height_sometimes = [0.7, 3.3]
ax.bar(bars, height = height_sometimes, bottom = height_always, label = 'sometimes', color = '#58508d')
height_never = [0.4, 1]
bottom_never = [99.6, 99]
ax.bar(bars, height = height_never, bottom = bottom_never, label = 'never', color = '#bc5090')
ax.set_facecolor('White')
plt.legend()
plt.xlabel('gender')
plt.ylabel('percent')
plt.title('Percent of Each Gender who Use Face Mask At Each Level (Brazil)')
```
