```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd

#size chart 
fig = plt.subplots(figsize=(12, 10))
# sales values of each group
mm = [97, 382.5, 947, 2010]
ans = [25.8, 68.9, 380.4, 519]
 
# Heights of mm + ans
bars = np.add(mm, ans).tolist()
 
# The position of the bars on the x-axis
r = [0,1,2,3]
 
# Names of group and bar width
names = ['Dec 30, 2018 - May 3, 2019', 'Apr 16 - May 3, 2019', 'Apr 16 - May 3, 2020', 'Dec 30, 2019 - May 3, 2020']
barWidth = 1

#blue bar - medical masks
plt.bar(r, mm, color='#58508d', edgecolor='white', width=barWidth, label = "Medical Masks")
#purple bar on top of blue - antiseptics
plt.bar(r, ans, bottom=mm, color='#bc5090', edgecolor='white', width=barWidth, label = "Antiseptics")
# Custom X axis
plt.xticks(r, names, fontsize = 9)
plt.yticks(np.arange(0,2501, 250))
plt.xlabel("Time Periods", fontsize = 20)
plt.ylabel("Sales in Millions of Russian Rubles", fontsize = 20)
plt.title("Sales Value of Medical Masks and Antiseptics in Russia", fontsize = 25)
# Show graphic
plt.legend()
plt.show()
```
